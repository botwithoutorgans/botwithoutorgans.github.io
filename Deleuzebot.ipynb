```python
import os
from google.cloud import translate_v2 as translate
import whisper

# Set up the Google Translate API client
os.environ["GOOGLE_APPLICATION_CREDENTIALS"] = "creds/credentials.json"
translate_client = translate.Client()

# Define the input directory and output directory
input_dir = "audio_files"
output_dir = "transcripts"

# Load the whisper model
model = whisper.load_model("base")

# Loop over the audio files and transcribe them
for filename in os.listdir(input_dir):
    if filename.endswith(".m4a") or filename.endswith(".wav"):
        audio_path = os.path.join(input_dir, filename)
        text = model.transcribe(audio_path)["text"]
        translated_text = translate_client.translate(text, target_language="en")["translatedText"]
        with open(os.path.join(output_dir, f"{filename}.txt"), "w") as f:
            f.write(translated_text)
```
